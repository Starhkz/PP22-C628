to get the correlation between missing valuesusing heat map
The missingno correlation heatmap measures nullity correlation: how strongly the presence or absence of one variable affects the presence of another


```python
msno.heatmap(df1)
```

Nullity correlation ranges from -1 (if one variable appears the other definitely does not) to 0 (variables appearing or not appearing have no effect on one another) to 1 (if one variable appears the other definitely also does).
1. it can be inferred from the heat map that classification features such as age, status, classification size, research intensity are highly uncorrelated with 0.1 values, that is features appearing or not appearing have no effect on one another 
2. Citations_per_Faculty_RANK, international_faculty_rank, the international students rank is directly correlated to the overall score giving a correlation of 1

```python
df1 = df1.iloc[:, [i for i, n in enumerate(np.var(df1.isnull(), axis='rows')) if n > 0]]
corr_mat = df1.isnull().corr()
print(corr_mat)
```

```python
#drop features not used for ranking  :Classification_size', 'Focus','Research Intensity' 'Age', 'Size']
to_drop =['Classification_size', 'Focus','Research Intensity', 'Age', 'Status']
df1_rankingfactors = df1.drop(to_drop, axis=1)
print (df1_rankingfactors)
```
